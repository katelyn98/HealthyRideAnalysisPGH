```python
import pandas as pd

from pathlib import Path

input_file = Path('.') / 'Bike Data'

f = list(input_file.glob("*.xlsx"))[0]

df = pd.read_excel(str(f))
print(df)
```

```python
new_df = df.rename(columns={"Trip id": "trip_id", "Starttime": "start_time", "Stoptime" : "stop_time", "Bikeid": "bike_id", "Tripduration": "trip_dur", "From station id": "from_station_id", "From station name": "from_station_name", "To station id": "to_station_id", "To station name": "to_station_name", "Usertype": "usertype"})
print(new_df)
df_filtered = new_df.query('from_station_id != None & to_station_id != None')
print(df_filtered)
```
